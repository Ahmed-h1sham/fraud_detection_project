```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

print("Imports loaded.")
```

```python
train_benef = pd.read_csv("../data/Train_Beneficiarydata.csv")
train_inp = pd.read_csv("../data/Train_Inpatientdata.csv")
train_outp = pd.read_csv("../data/Train_Outpatientdata.csv")
train_labels = pd.read_csv("../data/Train_labels.csv")

print("Training data loaded.")
```

```python
print("Beneficiary Data:")
display(train_benef.info())

print("\nInpatient Data:")
display(train_inp.info())

print("\nOutpatient Data:")
display(train_outp.info())

print("\nLabels:")
display(train_labels.info())
```

```python
date_cols_inp = ['AdmissionDt', 'DischargeDt', 'ClaimStartDt', 'ClaimEndDt']
for col in date_cols_inp:
    train_inp[col] = pd.to_datetime(train_inp[col], errors='coerce')

date_cols_out = ['ClaimStartDt', 'ClaimEndDt']
for col in date_cols_out:
    train_outp[col] = pd.to_datetime(train_outp[col], errors='coerce')

train_benef['DOB'] = pd.to_datetime(train_benef['DOB'], errors='coerce')
train_benef['DOD'] = pd.to_datetime(train_benef['DOD'], errors='coerce')

print("Date conversion complete.")
```

```python
reference_date = pd.to_datetime("2009-12-01")
train_benef["Age"] = (reference_date - train_benef["DOB"]).dt.days // 365

train_benef[["BeneID","DOB","Age"]].head()
```

```python
train_inp["LOS"] = (train_inp["DischargeDt"] - train_inp["AdmissionDt"]).dt.days
train_inp[["AdmissionDt", "DischargeDt", "LOS"]].head()
```

```python
train_inp = train_inp.rename(columns={"Provider": "ProviderID"})
train_outp = train_outp.rename(columns={"Provider": "ProviderID"})
train_labels = train_labels.rename(columns={"Provider": "ProviderID"})

print("Renaming complete. Columns fixed:")
print(train_inp.columns[:10])
print(train_outp.columns[:10])
print(train_labels.columns)
```

```python
inp_merged = train_inp.merge(train_benef, on="BeneID", how="left")
outp_merged = train_outp.merge(train_benef, on="BeneID", how="left")

print("Merging complete.")
```

```python
inp_provider = inp_merged.groupby("ProviderID").agg({
    "LOS": ["mean", "max", "min"],
    "InscClaimAmtReimbursed": ["mean", "sum"],
    "ClaimID": "count"
})

inp_provider.columns = ["INP_" + "_".join(col) for col in inp_provider.columns]
inp_provider.head()
```

```python
outp_provider = outp_merged.groupby("ProviderID").agg({
    "InscClaimAmtReimbursed": ["mean", "sum"],
    "ClaimID": "count"
})

outp_provider.columns = ["OUT_" + "_".join(col) for col in outp_provider.columns]
outp_provider.head()
```

```python
provider_features = inp_provider.join(outp_provider, how="outer").fillna(0)
provider_features.head()
```

```python
provider_features = provider_features.merge(train_labels, on="ProviderID", how="left")
provider_features.head()
```

```python
provider_features.info()
provider_features["PotentialFraud"].value_counts(dropna=False)
```

```python
sns.countplot(data=provider_features, x="PotentialFraud")
plt.title("Fraud vs Non-Fraud Provider Distribution")
plt.show()
```

```python
provider_features.to_csv("provider_features_final.csv", index=False)
print("Saved provider_features_final.csv")
```
